Neste projeto, utilizaremos métodos numéricos baseados em Álgebra Linear Computacional para solucionar o problema de remoção de fundo em vídeos de câmeras de segurança. Especificamente, faremos uso do método de Análise de Componentes Principais Robusta (Robust Principal Component Analysis - PCA) para realizar essa tarefa.

O método de PCA é uma técnica amplamente utilizada em Álgebra Linear e Processamento de Sinais para analisar e reduzir a dimensionalidade de conjuntos de dados. Ele é útil para identificar as principais características (componentes principais) presentes nos dados e expressar os dados originais em termos dessas componentes principais. Isso permite uma representação mais compacta e informativa dos dados.

No contexto da remoção de fundo em vídeos, o PCA robusto se mostra especialmente útil. Ele é capaz de lidar com variações complexas e não-lineares nas imagens, incluindo oclusões, ruído e mudanças de iluminação. Ao decompor o vídeo em suas componentes principais, podemos separar o fundo (componentes de baixa variação) dos objetos em movimento (componentes de alta variação).

O PCA robusto é capaz de identificar o fundo estático e eliminar as variações de curto prazo (movimento) presentes nas imagens, fornecendo uma representação do vídeo apenas com as informações relevantes dos objetos em movimento. Essa abordagem permite isolar as pessoas que transitam no ambiente, facilitando sua detecção e rastreamento posterior.

Ao aplicar o PCA robusto aos vídeos de câmeras de segurança em um comércio, podemos remover o fundo estático, eliminando o ruído e as variações indesejadas, e obter uma representação mais limpa e concentrada nas pessoas em movimento. Isso melhora a precisão dos algoritmos de detecção e rastreamento, permitindo identificar e acompanhar as pessoas de maneira mais eficiente.

Portanto, a escolha do PCA robusto como método numérico nesse projeto se deve à sua capacidade de lidar com as características específicas do problema, como variações complexas, oclusões e ruído. Ele nos permite separar o fundo do vídeo e concentrar-se nas informações relevantes para a identificação das pessoas, melhorando a qualidade e a precisão dos resultados finais.

```python
#função para converter para a escala de cinza
def rgb2gray(rgb): 
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

#checagem dos frames
def check_frames(clip, frames): 
    num_frames = clip.fps * clip.duration
    desired_dur = frames / clip.fps 
    assert desired_dur <= clip.duration, 'muitos quadros solicitados! deve ser menor que {}'.format(num_frames)

    #Crie uma matriz numpy a partir do clipe de vídeo
def video_to_matrix(clip, frames=20, scale=0.10):
    check_frames(clip, frames)
    image_list = []
    for frame in range(frames):
        time = frame/clip.fps
        gray = rgb2gray(clip.get_frame(time)).astype(int)
        dims = (1280,720)
        resized = resize(gray, dims, anti_aliasing=True)
        flattened = resized.flatten() 
        image_list.append(flattened)
    return np.vstack(image_list).T
```

```python
scale = 1
dims  = (int(1280), int(720)) 
```

```python
#convertendo o video em uma matriz númerica
video_mat = video_to_matrix(video, frames=100, scale=scale)
print('dims: {} \nvideo matrix dimensões {}'.format(dims, video_mat.shape))
```

### PCA (Principal Component Analysis) 

```python
#fazendo a decomposição SVD
U, S, Vt = decomposition.randomized_svd(video_mat, 10)
print (U.shape, S.shape, Vt.shape)
```

```python
#obtendo a low_rank 
low_rank = U @ np.diag(S) @ Vt
low_rank.shape
```

 A matriz de baixo rank contém as informações principais e mais significativas dos dados, enquanto os componentes de alta magnitude nos valores singulares (presentes na matriz S) correspondem aos principais modos de variação.

```python
plt.figure(figsize=(12, 12))
plt.imshow(low_rank, aspect="auto", cmap='gray')
```

```python
SVD_bg_removido = video_mat-low_rank #removendo a matriz de baixa energia
```

```python
plot_duas_imagens(video_mat[:,20].reshape(dims),
                SVD_bg_removido[:, 20].reshape(dims), cmap='gray')
```

```python
SVD_clipped = np.clip(SVD_bg_removido, None, 10)

plot_duas_imagens(video_mat[:,20].reshape(dims),
                SVD_clipped[:, 20].reshape(dims), cmap='gray')
```

```python
#removendo background usando SVD 
def SVD_background_removal(clip, frames=5, scale=1, n_components=2):
    #converte um video clip em um video tensor com SVD 
    
    
    #SVD
    A = video_to_matrix(clip, frames=frames, scale=scale) # convertendo o video em matriz
    U, s, Vh = decomposition.randomized_svd(A, n_components) #decomposição SVD
    low_rank = U @ np.diag(s) @ Vh #obtendo a matriz de baixo rank
    
    #remove background 
    bg_removed = A-low_rank
    bg_removed = bg_removed.reshape((960, 960, 100))
    return bg_removed
```

```python
bg = SVD_background_removal(video, frames=100)
bg.shape
```

```python
fps = video.fps
fig, ax = plt.subplots()
#criando o video sem background
def make_frame(t):
    ax.clear()
    
    #processando o frame 
    img = bg[:, : ,int(t*fps) % 100 ]
    
    #removendo ruido
    mean = np.mean(img)
    std = np.std(img)
    threshold = mean + 3 * std  
    img = np.clip(img, None, threshold)
    
    #output 
    ax.imshow(img, cmap='gray')
    out = mplfig_to_npimage(fig)[:720, :1280, :100]
    out = Image.fromarray(np.uint8(out))
    out = np.array(out)
    return out 


animation = mpy.VideoClip(make_frame, duration=video.duration)
#escrevendo o video 
animation.write_videofile('C:/Users/Admin/Downloads/videoPessoasSemfundo.mp4', fps=fps)
```
